```python
df_combined
```

#### Ajout de variables de croissance

```python
df_combined = df_combined.sort_values(by=["Region", "Year"])

# Croissance absolue annuelle
df_combined["Yearly_Growth"] = df_combined.groupby("Region")["Internet_Users_Million"].diff().fillna(0)

# Croissance du taux de pénétration
df_combined["Yearly_Growth_Percent"] = df_combined.groupby("Region")["Internet_Penetration_Percent"].diff().fillna(0)
df_combined["Estimated_Population"] = df_combined["Internet_Users_Million"] / (df_combined["Internet_Penetration_Percent"] / 100)
```

```python
df_combined
```

#### Définition des features (X) et target (Y) et justification

##### Les features représentent les facteurs explicatifs, c’est-à-dire les variables que j’utilise pour prédire la valeur cible.
##### La target est la variable à estimer à partir des features.

#### Séparer les deux me permet :

#####   - d'entraîner les modèles sans jamais leur "donner la réponse" (la target),

#####   - de structurer mes données selon la logique input → output des modèles supervisés.

```python
df_prepared = pd.get_dummies(df_combined, columns=["Region"], drop_first=True)


features = [
    "Year",
    "Internet_Penetration_Percent",
    "Yearly_Growth",
    "Yearly_Growth_Percent",
    "Estimated_Population"
] + [col for col in df_prepared.columns if col.startswith("Region_")]

X = df_prepared[features]
y = df_prepared["Internet_Users_Million"]
```

### Justification des variables sélectionnées (features)

#### J’ai inclus dans mes features :

#####   1. Year : pour intégrer la tendance temporelle (l’évolution naturelle au fil des années).

#####   2. Internet_Penetration_Percent : donne une idée du taux d’adoption dans la population. Très corrélé au volume d’utilisateurs.

#####   3. Yearly_Growth : indicateur de dynamique de croissance, utile pour identifier les zones en développement rapide.

#####   4. Yearly_Growth_Percent : complémentaire, permet de détecter des changements structurels.

#####   5. Estimated_Population : estimée à partir du volume et du taux de pénétration, elle me permet d’avoir une idée de la taille potentielle du marché numérique.

#####   6. Region_... (one-hot encodées) : pour que les modèles puissent différencier les régions, sans introduire d’ordre arbitraire.



### Pourquoi inclure les variables de croissance ?

#### Les variables de croissance (Yearly_Growth, Yearly_Growth_Percent) capturent la variation d’une année à l’autre, ce que les valeurs brutes ne permettent pas de voir directement.

### Elles sont précieuses pour :

#####    détecter des accélérations ou ralentissements dans une région,

#####    donner aux modèles une vision dynamique, et pas juste une photo figée,

#####    améliorer la précision des prédictions sur des séries temporelles.

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.decomposition import PCA

# On va garder 2 composantes principales
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
```

```python
explained_var = pca.explained_variance_ratio_
print(f"Variance expliquée par la PC1 : {explained_var[0]:.2%}")
print(f"Variance expliquée par la PC2 : {explained_var[1]:.2%}")
print(f"Total conservé : {sum(explained_var):.2%}")
```

```python
import matplotlib.pyplot as plt

# On ajoute la région pour la couleur
import pandas as pd
df_pca = pd.DataFrame(X_pca, columns=["PC1", "PC2"])
df_pca["Region"] = df_combined["Region"].values

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_pca, x="PC1", y="PC2", hue="Region", palette="Set2")
plt.title("Projection des données dans le plan des deux premières composantes principales")
plt.grid(True)
plt.show()
```

#### La PCA montre que les deux premières composantes capturent environ 49 % de la variance.
#### Elles permettent de visualiser clairement la singularité de la région Asie-Pacifique, qui domine fortement en termes de volume.
#### Les autres régions sont plus proches, ce qui suggère une structure numérique plus homogène.
#### La projection est utile pour repérer des similarités ou des différences globales, mais une modélisation complète devra réutiliser toutes les variables originales, car près de 50 % de l’information reste en dehors du plan PC1-PC2.

```python
# On récupère les noms de colonnes utilisées dans X
feature_names = X.columns

# Les composantes de PCA (loadings)
loadings = pd.DataFrame(pca.components_.T, columns=['PC1', 'PC2'], index=feature_names)

# Afficher les poids triés pour PC1 et PC2
print("PC1 - Variables les plus influentes :")
print(loadings["PC1"].sort_values(ascending=False))

print("\nPC2 - Variables les plus influentes :")
print(loadings["PC2"].sort_values(ascending=False))
```

```python
#PC1 représente une dimension “taille du marché numérique”.
#Elle capte des différences structurelles massives, en particulier celles liées à la population estimée, 
#la croissance absolue du nombre d’utilisateurs, et le poids de l’Asie-Pacifique.
#Elle sépare clairement les régions massivement connectées (Asie-Pacifique) des autres.

#PC2 est une dimension “maturité numérique”.
#Elle capte les progrès temporels et qualitatifs des régions, notamment via :
#L’augmentation du taux de pénétration,
#La progression chronologique (Year)
#Et une contribution notable de l’Europe.

#Elle distingue plutôt les marchés matures (Europe, Amériques) des zones en développement (Afrique, Arab States).
```

#### Ces résultats justifient pleinement l’intérêt de la PCA dans ce projet : elle permet de résumer visuellement et numériquement des différences structurelles entre régions, sans avoir besoin de toutes les variables initiales

```python
df_pca = pd.DataFrame(X_pca, columns=["PC1", "PC2"])
df_pca["Region"] = df_combined["Region"].values

# Taille du graphe
plt.figure(figsize=(12, 7))

# Scatter plot des observations
sns.scatterplot(data=df_pca, x="PC1", y="PC2", hue="Region", palette="Set2", alpha=0.7, s=60)

# Ajouter les vecteurs des variables (loadings)
for i, var in enumerate(loadings.index):
    x = loadings.iloc[i, 0]  # PC1
    y = loadings.iloc[i, 1]  # PC2
    plt.arrow(0, 0, x*3, y*3, color='black', alpha=0.6, head_width=0.05)
    plt.text(x*3.2, y*3.2, var, fontsize=10, color='black')

# Titres et style
plt.title("Biplot PCA : projection des données et vecteurs des variables")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.grid(True)
plt.axhline(0, color='grey', linewidth=0.5)
plt.axvline(0, color='grey', linewidth=0.5)
plt.tight_layout()
plt.show()
```

### Modélisation avec SVM en régression

```python
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error, r2_score
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)
kernels = ['linear', 'poly', 'rbf', 'sigmoid']
results = []

for kernel in kernels:
    model = SVR(kernel=kernel, C=1.0, epsilon=0.1)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    
    results.append({
        'Kernel': kernel,
        'MSE': mse,
        'R2': r2
    })

results_df = pd.DataFrame(results)
print(results_df.sort_values(by="R2", ascending=False))
```

### Optimisation performances de SVM linear via GridSearchCV (tuning des paramètres)

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVR

# 1. Définir le modèle de base
svm = SVR(kernel='linear')

# 2. Définir la grille d'hyperparamètres à tester
param_grid = {
    'C': [0.1, 1, 10, 100],
    'epsilon': [0.01, 0.1, 0.5, 1.0]
}

# 3. GridSearch avec validation croisée
grid_search = GridSearchCV(
    estimator=svm,
    param_grid=param_grid,
    scoring='r2',
    cv=5,  # 5-fold cross-validation
    n_jobs=-1,  # utilise tous les cœurs dispo
    verbose=1
)

# 4. Entraînement sur le jeu complet d'entraînement
grid_search.fit(X_train, y_train)

# 5. Résultats
print("✅ Meilleurs paramètres :", grid_search.best_params_)
print(f"✅ Score R² sur validation croisée : {grid_search.best_score_:.2f}")
```

### On choisis un kernel linear pour son R2 plus élévé et on prend les paramètres données par GridSearchCV

```python
from sklearn.svm import SVR
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt

# 1. Redéfinir la cible si nécessaire
y = df_prepared["Internet_Users_Million"]

# 2. Split des données
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)

# 3. Modèle SVM avec noyau linéaire
svm_linear = SVR(kernel='linear', C=100, epsilon=0.01)
svm_linear.fit(X_train, y_train)

# 4. Prédictions
y_pred = svm_linear.predict(X_test)

# 5. Évaluation
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"🌟 SVM (linéaire) - MSE : {mse:.2f}")
print(f"🌟 SVM (linéaire) - R² : {r2:.2f}")
```

#### Visualisation graphique

```python
plt.figure(figsize=(8, 5))
plt.scatter(y_test, y_pred, color='green', alpha=0.6)
plt.plot([y.min(), y.max()], [y.min(), y.max()], 'k--', lw=2)
plt.xlabel("Valeurs réelles")
plt.ylabel("Prédictions SVM (linéaire)")
plt.title("SVM linéaire - Prédictions vs Réalité")
plt.grid(True)
plt.tight_layout()
plt.show()
```

##### Grâce au GridSearchCV, j’ai optimisé les hyperparamètres C et epsilon du SVM linéaire.
##### Le score R² est passé de 0.43 à 0.79, ce qui montre que le modèle explique désormais près de 80 % de la variance du nombre d’utilisateurs d’Internet.
##### Le modèle généralisé est beaucoup plus précis, comme l’illustre le graphe de prédictions.
##### Ce tuning est indispensable pour exploiter tout le potentiel du SVM.

### Decision Tree 

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error, r2_score

# Modèle de base
tree = DecisionTreeRegressor(random_state=42)
tree.fit(X_train, y_train)

# Prédiction
y_pred_tree = tree.predict(X_test)

# Évaluation
mse_tree = mean_squared_error(y_test, y_pred_tree)
r2_tree = r2_score(y_test, y_pred_tree)

print(f"🌳 Decision Tree - MSE : {mse_tree:.2f}")
print(f"🌳 Decision Tree - R² : {r2_tree:.2f}")
```

```python
# Conversion en DataFrame
importances = pd.Series(tree.feature_importances_, index=X.columns).sort_values(ascending=True)
importance_df = pd.DataFrame({
    "Feature": importances.index,
    "Importance": importances.values
})

plt.figure(figsize=(10, 6))
sns.barplot(
    data=importance_df,
    x="Importance",
    y="Feature"
)

plt.title("Importance des variables - Decision Tree")
plt.xlabel("Importance relative")
plt.ylabel("Variables")
plt.grid(True, axis='x', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

## GridSearchCV – Decision Tree Regressor

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import GridSearchCV

# Grille d’hyperparamètres à tester
param_grid_tree = {
    'max_depth': [3, 5, 10, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

# Instanciation GridSearchCV
grid_tree = GridSearchCV(
    DecisionTreeRegressor(random_state=42),
    param_grid=param_grid_tree,
    scoring='r2',
    cv=5,
    n_jobs=-1,
    verbose=1
)

# Entraînement
grid_tree.fit(X_train, y_train)

# Résultats
print("🌳 Meilleurs paramètres - Decision Tree :", grid_tree.best_params_)
print(f"🌳 Score R² (val. croisée) : {grid_tree.best_score_:.2f}")

# Évaluation finale
best_tree = grid_tree.best_estimator_
y_pred_tree_best = best_tree.predict(X_test)

from sklearn.metrics import mean_squared_error, r2_score
mse_tree_best = mean_squared_error(y_test, y_pred_tree_best)
r2_tree_best = r2_score(y_test, y_pred_tree_best)

print(f"🌳 Decision Tree Optimisé - MSE : {mse_tree_best:.2f}")
print(f"🌳 Decision Tree Optimisé - R² : {r2_tree_best:.2f}")
```

## GridSearchCV - Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor

# Grille d’hyperparamètres
param_grid_rf = {
    'n_estimators': [50, 100, 200],
    'max_depth': [5, 10, None],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2]
}

grid_rf = GridSearchCV(
    RandomForestRegressor(random_state=42),
    param_grid=param_grid_rf,
    scoring='r2',
    cv=5,
    n_jobs=-1,
    verbose=1
)

grid_rf.fit(X_train, y_train)

print("🌲 Meilleurs paramètres - Random Forest :", grid_rf.best_params_)
print(f"🌲 Score R² (val. croisée) : {grid_rf.best_score_:.2f}")

best_rf = grid_rf.best_estimator_
y_pred_rf_best = best_rf.predict(X_test)

mse_rf_best = mean_squared_error(y_test, y_pred_rf_best)
r2_rf_best = r2_score(y_test, y_pred_rf_best)

print(f"🌲 Random Forest Optimisé - MSE : {mse_rf_best:.2f}")
print(f"🌲 Random Forest Optimisé - R² : {r2_rf_best:.2f}")
```
